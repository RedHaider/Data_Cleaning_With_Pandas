```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = 'https://dsebd.org/company_listing.php'
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the Scroller div element
scroller = soup.find('div', {'class': 'Scroller'})

# Find all the scroll-item div elements within the Scroller
scroll_items = scroller.find_all('div', {'class': 'scroll-item'})

# Iterate over the scroll items and extract the company details
for item in scroll_items:
    table = item.find('table')
    rows = table.find_all('tr')
    for row in rows:
        link = row.find('a')
        if link:
            company_link = link['href']
            code = company_link.split('=')[-1]
            name = link.text.strip()
            print(f"Name: {name}")
            print(f"Code: {code}")
            print(f"Link: {company_link}\n")
```

```python
import csv
# Create a list to store the company details
companies = []

# Iterate over the scroll items and extract the company details
for item in scroll_items:
    table = item.find('table')
    rows = table.find_all('tr')
    for row in rows:
        link = row.find('a')
        if link:
            company_link = link['href']
            code = company_link.split('=')[-1]
            name = link.text.strip()
            companies.append({
                'Name': name,
                'Code': code,
                'Link': company_link
            })

# Specify the CSV file path
csv_file = '/content/drive/MyDrive/ADN_DATA/exam_1/company_details.csv'

# Write the company details to the CSV file
with open(csv_file, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=['Name', 'Code', 'Link'])
    writer.writeheader()
    writer.writerows(companies)

print(f"Company details saved to {csv_file} successfully.")
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://dsebd.org/displayCompany.php?name=AAMRANET'

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table that contains the company details
table = soup.find('table', {'class': 'table'})

# Find the rows in the table
rows = table.find_all('tr')

# Iterate over the rows and extract the company details
company_details = {}
for row in rows:
    columns = row.find_all('td')
    if len(columns) == 2:
        label = columns[0].text.strip()
        value = columns[1].text.strip()
        company_details[label] = value

# Print the extracted company details
print('Company Name:', company_details['Company Name'])
print('Trading Code:', company_details['Trading Code'])
print('Scrip Code:', company_details['Scrip Code'])
print('Sector:', company_details['Sector'])
```
