```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = 'https://dsebd.org/company_listing.php'
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the Scroller div element
scroller = soup.find('div', {'class': 'Scroller'})

# Find all the scroll-item div elements within the Scroller
scroll_items = scroller.find_all('div', {'class': 'scroll-item'})

# Iterate over the scroll items and extract the company details
for item in scroll_items:
    table = item.find('table')
    rows = table.find_all('tr')
    for row in rows:
        link = row.find('a')
        if link:
            company_link = link['href']
            code = company_link.split('=')[-1]
            print(f"Code: {code}")
            print(f"Link: {company_link}\n")
```

```python
import csv
# Create a list to store the company details
companies = []

# Iterate over the scroll items and extract the company details
for item in scroll_items:
    table = item.find('table')
    rows = table.find_all('tr')
    for row in rows:
        link = row.find('a')
        if link:
            company_link = link['href']
            code = company_link.split('=')[-1]
            companies.append({
                'Code': code,
                'Link': company_link
            })

# Specify the CSV file path
csv_file = '/content/drive/MyDrive/ADN_DATA/exam_1/company_details.csv'

# Write the company details to the CSV file
with open(csv_file, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=['Code', 'Link'])
    writer.writeheader()
    writer.writerows(companies)

print(f"Company details saved to {csv_file} successfully.")
```

```python
import requests
from bs4 import BeautifulSoup

# Send GET request to the webpage
url = "https://dsebd.org/displayCompany.php?name=AAMRANET"
response = requests.get(url)
html_content = response.content

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find the company name
company_name_element = soup.find('h2', class_='BodyHead topBodyHead')
company_name = company_name_element.text.strip() if company_name_element else "Not Found"

# Find the trading code and scrip code
data_element = soup.find('table', class_='table-bordered')
if data_element:
    trading_code_element = data_element.find('strong', string='Trading Code:')
    scrip_code_element = data_element.find('strong', string='Scrip Code:')
    if trading_code_element:
        trading_code = trading_code_element.next_sibling.strip()
    else:
        trading_code = "Not Found"
    if scrip_code_element:
        scrip_code = scrip_code_element.next_sibling.strip()
    else:
        scrip_code = "Not Found"
else:
    trading_code = "Not Found"
    scrip_code = "Not Found"

# Find the sector
sector_element = soup.find('th', string='Sector')
sector = sector_element.find_next_sibling('td').text.strip() if sector_element else "Not Found"

# Print the scraped data
print("Company Name:", company_name)
print("Trading Code:", trading_code)
print("Scrip Code:", scrip_code)
print("Sector:", sector)
```

```python
import csv
import pandas as pd
import requests
from bs4 import BeautifulSoup

# CSV file path
csv_file = '/content/drive/MyDrive/ADN_DATA/exam_1/company_details.csv'
# Output files
output_csv_file = '/content/drive/MyDrive/ADN_DATA/exam_1/scraped_data.csv'
output_excel_file = '/content/drive/MyDrive/ADN_DATA/exam_1/scraped_data.xlsx'

# Send GET request and parse HTML content for each link
data = []
header = ["Code", "Company Name", "Trading Code", "Scrip Code", "Sector"]
data.append(header)

with open(csv_file, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header row
    for row in csv_reader:
        code = row[0]
        link = "https://dsebd.org/" + row[1]

        # Send GET request to the webpage
        response = requests.get(link)
        html_content = response.content

        # Parse the HTML content
        soup = BeautifulSoup(html_content, 'html.parser')

        # Find the company name
        company_name_element = soup.find('h2', class_='BodyHead topBodyHead')
        company_name = company_name_element.text.strip() if company_name_element else "Not Found"

        # Find the trading code and scrip code
        data_element = soup.find('table', class_='table-bordered')
        if data_element:
            trading_code_element = data_element.find('strong', string='Trading Code:')
            scrip_code_element = data_element.find('strong', string='Scrip Code:')
            if trading_code_element:
                trading_code = trading_code_element.next_sibling.strip()
            else:
                trading_code = "Not Found"
            if scrip_code_element:
                scrip_code = scrip_code_element.next_sibling.strip()
            else:
                scrip_code = "Not Found"
        else:
            trading_code = "Not Found"
            scrip_code = "Not Found"

        # Find the sector
        sector_element = soup.find('th', string='Sector')
        sector = sector_element.find_next_sibling('td').text.strip() if sector_element else "Not Found"

        # Append the scraped data to the list
        data.append([code, company_name, trading_code, scrip_code, sector])

# Save the data to a CSV file
with open(output_csv_file, 'w', newline='') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerows(data)

# Save the data to an Excel file
df = pd.DataFrame(data[1:], columns=data[0])
df.to_excel(output_excel_file, index=False)
```
